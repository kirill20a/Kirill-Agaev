```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_excel('123.xlsx', sheet_name='Товары')
```

```python
df['Дата продажи'] = pd.to_datetime(df['Дата продажи'])
```

```python
df['Месяц'] = df['Дата продажи'].dt.month
```

```python
df['Квартал'] = df['Дата продажи'].dt.quarter
```

```python
df['Выручка'] = df['Кол-во'] * df['Стоимость']
```

# 1. Сводная таблица: Общая выручка по группам товаров и магазинам

```python
print("1. ОБЩАЯ ВЫРУЧКА ПО ГРУППАМ ТОВАРОВ И МАГАЗИНАМ")
```

```python
pivot1 = df.pivot_table(
    values='Выручка',
    index='Группа товаров',
    columns='Название магазина',
    aggfunc='sum',
    fill_value=0,
    margins=True,
    margins_name='ВСЕГО'
)
```

```python
pivot1_formatted = pivot1.applymap(lambda x: f"{x:,.0f} руб." if pd.notnull(x) else "0 руб.")
```

```python
print(pivot1_formatted)
```

# 2. Сводная таблица: Средний чек и количество продаж по месяцам

```python
print("2. СРЕДНИЙ ЧЕК И КОЛИЧЕСТВО ПРОДАЖ ПО МЕСЯЦАМ")
```

```python
pivot2 = df.pivot_table(
    values=['Выручка', 'Кол-во'],
    index='Месяц',
    aggfunc={'Выручка': 'sum', 'Кол-во': 'sum'},
    margins=True,
    margins_name='ВСЕГО'
)
```

```python
pivot2['Средний чек'] = pivot2['Выручка'] / pivot2['Кол-во']
pivot2['Выручка'] = pivot2['Выручка'].apply(lambda x: f"{x:,.0f} руб.")
pivot2['Средний чек'] = pivot2['Средний чек'].apply(lambda x: f"{x:,.0f} руб.")
pivot2['Кол-во'] = pivot2['Кол-во'].apply(lambda x: f"{x:,.0f} шт.")
```

```python
print(pivot2)
```

# 3. Сводная таблица: Топ-5 самых продаваемых товаров по выручке

```python
print("3. ТОП-5 САМЫХ ПРОДАВАЕМЫХ ТОВАРОВ ПО ВЫРУЧКЕ")
```

```python
top_products = df.groupby('Название').agg({
    'Выручка': 'sum',
    'Кол-во': 'sum',
    'Стоимость': 'mean',
    'Группа товаров': 'first'
}).nlargest(5, 'Выручка')
```

```python
top_products['Выручка'] = top_products['Выручка'].apply(lambda x: f"{x:,.0f} руб.")
top_products['Стоимость'] = top_products['Стоимость'].apply(lambda x: f"{x:,.0f} руб.")
top_products['Кол-во'] = top_products['Кол-во'].apply(lambda x: f"{x:,.0f} шт.")
```

```python
print(top_products)
```

# 4. Сводная таблица: Динамика продаж по кварталам и группам товаров

```python
print("4. ДИНАМИКА ПРОДАЖ ПО КВАРТАЛАМ И ГРУППАМ ТОВАРОВ")
```

```python
pivot4 = df.pivot_table(
    values='Выручка',
    index='Группа товаров',
    columns='Квартал',
    aggfunc='sum',
    fill_value=0
)
```

```python
pivot4['Итого'] = pivot4.sum(axis=1)
pivot4 = pivot4.sort_values('Итого', ascending=False)
```

```python
pivot4_formatted = pivot4.applymap(lambda x: f"{x:,.0f} руб." if pd.notnull(x) else "0 руб.")
```

```python
print(pivot4_formatted)
```

# 5. Сводная таблица: Эффективность магазинов (выручка на 1 магазин)

```python
print("5. ЭФФЕКТИВНОСТЬ МАГАЗИНОВ")
```

```python
store_stats = df.pivot_table(
    values=['Выручка', 'Кол-во'],
    index='Название магазина',
    aggfunc={'Выручка': ['sum', 'count'], 'Кол-во': 'sum'}
)
```

```python
store_stats.columns = ['Количество_продаж', 'Общая_выручка', 'Общее_количество']
store_stats['Выручка_на_одну_продажу'] = store_stats['Общая_выручка'] / store_stats['Количество_продаж']
store_stats['Среднее_количество_в_продаже'] = store_stats['Общее_количество'] / store_stats['Количество_продаж']
```

```python
store_stats['Общая_выручка'] = store_stats['Общая_выручка'].apply(lambda x: f"{x:,.0f} руб.")
store_stats['Выручка_на_одну_продажу'] = store_stats['Выручка_на_одну_продажу'].apply(lambda x: f"{x:,.0f} руб.")
store_stats['Среднее_количество_в_продаже'] = store_stats['Среднее_количество_в_продаже'].apply(lambda x: f"{x:.1f} шт.")
store_stats['Общее_количество'] = store_stats['Общее_количество'].apply(lambda x: f"{x:,.0f} шт.")
```

```python
print(store_stats)
```
